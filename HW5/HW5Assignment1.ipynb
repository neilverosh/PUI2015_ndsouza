A second degree curve is fitted to the regression of Total Units vs Total Energy Consumption.

##Inference from plots

The second degree curve is the best fit as seen by its high R-squared value. The other two models have the the same R-squared value and hence cannot be compared on the basis of fit. 
Log-likelihood cannot be computed for any of the models since they all have the same degrees of freedom (Df Model = 1). A difference in degrees of freedom of two models is needed as log-likehood ratio (statistic) takes a chi-squared distribution.
