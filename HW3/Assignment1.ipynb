```python
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np
%matplotlib inline
import json
s = json.load( open('plotformat.json') )
plt.rcParams.update(s)
```

```python
mymean = 100
```

```python
mysizes = 2000/(np.array(range(1,100)+[10]))
#mysizes = np.random.choice((range(10,2000)), 100, replace = False)
```

```python
distributions ={'Normal':{}, 'Poisson':{}, 'Binomial':{}, 'ChiSq':{}, 'Laplace':{}, 'Rayleigh':{}}
for n in mysizes:
    distributions['Normal'][n] = np.random.normal(mymean, 1, n)
    distributions['Poisson'][n] = np.random.poisson(mymean, n)
    distributions['Binomial'][n] = np.random.binomial(200, 0.5, n)
    distributions['ChiSq'][n] = np.random.chisquare(mymean, n)
    distributions['Laplace'][n] = np.random.laplace(mymean, 1, n)
    distributions['Rayleigh'][n] = np.random.rayleigh((np.sqrt(2 / np.pi) * mymean), n)
```

```python
means = {}
for i in distributions:
    means[i] = {}
    fig_mu_N = plt.figure(figsize=(10,10))
    axi_mu_n = fig_mu_N.add_subplot(1,1,1)
    for nn in distributions[i].iterkeys():
        means[i][nn] = distributions[i][nn].mean()
        axi_mu_n.plot(nn, means[i][nn].mean(), 'o')
        axi_mu_n.set_xlabel('sample size')
        axi_mu_n.set_ylabel('sample mean')
        axi_mu_n.set_title(i)
        axi_mu_n.plot([min(mysizes), max(mysizes)], [mymean,mymean], 'k')
```

```python
allmeans=[]
for i in distributions:
    for n in means[i]:
        allmeans.append(means[i][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
plt.title('Histogram of sample means')
```

In general most of these plots assume a shape close to gaussion (normal) distribution which supports the central limit theorum.
